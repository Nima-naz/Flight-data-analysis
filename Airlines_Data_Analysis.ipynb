### 4. Key Insights & Findings

After conducting extensive Exploratory Data Analysis (EDA) on the airline flight dataset, several important patterns and relationships were identified.  
Below are the most significant findings derived from the visualizations:

---


### ✈️ 1. Ticket Pricing Trends
- **Ticket prices are right-skewed**, meaning most fares are in the lower range, but a few flights are significantly more expensive.
- **Business class** fares are considerably higher and more consistent, while **Economy class** fares show a wide spread.
- **Non-stop flights** are generally more expensive than one- or two-stop flights.
---

### 📅 2. Booking Window Impact
- **Prices increase sharply as the number of days left decreases.**
  - The line chart shows a steep upward curve for last-minute bookings (0–2 days before departure).
- **Booking early** (60–90 days in advance) consistently leads to lower fares.
- This confirms that airlines use **dynamic pricing** based on time to departure.

---

### 🏢 3. Airline-wise Price Comparison
- Premium carriers like **Vistara** and **Air India** have **higher median prices** across routes and classes.
- Budget airlines such as **SpiceJet** and **AirAsia** offer cheaper average fares.
- The variability within each airline suggests active **revenue management and route-based pricing strategies.**
---

### 🕒 4. Departure Time Influence
- **Morning and afternoon flights** tend to be more expensive than evening or night flights.
- **Early Morning** and **Late Night** flights offer cheaper fares, possibly due to lower demand.
- This indicates a **time-of-day pricing pattern** reflecting customer convenience and demand cycles.

---

### 🧳 5. Route and Duration Analysis
- Popular routes like **Delhi–Mumbai** and **Bangalore–Delhi** consistently have higher average prices due to heavy demand.
- **Longer flight durations** correlate positively with price, but the relationship is **not perfectly linear** — other factors such as airline and stops play a role.
- **Short-haul flights** sometimes have unexpectedly high prices if operated by premium airlines or during peak times.

---


### 🌍 6. City and Class Distributions
- **Delhi, Mumbai, Bangalore, and Hyderabad** emerge as major travel hubs — both as sources and destinations.
- Around **70–75% of tickets** are in **Economy class**, showing its dominance in the domestic airline market.
- **Business class** makes up the remaining share, serving a niche segment with higher fare stability.

---

### 🔍 7. Correlation and Overall Patterns
- The **correlation matrix** indicates:
  - A **negative correlation between days_left and price**, confirming the “book early, pay less” trend.
  - A **positive correlation between duration and price**, as longer flights generally cost more.
- These relationships validate standard airline pricing logic and confirm the data’s consistency.

---


### 🧠 Overall Summary
- The dataset reveals **strong pricing dynamics** influenced by airline, class, booking time, stops, and route.  
- **Early booking** and **choosing non-peak hours or connecting flights** are the best strategies for lower fares.  
- Airlines maintain competitive differentiation via **route-based and time-based pricing** to maximize profitability.
